```python
import cv2

# Define a dictionary to map the property constants to their names
prop_id_to_name = {
    cv2.CAP_PROP_POS_MSEC: 'Current position in milliseconds',
    cv2.CAP_PROP_POS_FRAMES: 'Current position in frames',
    cv2.CAP_PROP_POS_AVI_RATIO: 'Relative position of the video file',
    cv2.CAP_PROP_FRAME_WIDTH: 'Width of the frames in the video stream',
    cv2.CAP_PROP_FRAME_HEIGHT: 'Height of the frames in the video stream',
    cv2.CAP_PROP_FPS: 'Frame rate',
    cv2.CAP_PROP_FOURCC: '4-character code of codec',
    cv2.CAP_PROP_FRAME_COUNT: 'Number of frames in the video file',
    cv2.CAP_PROP_FORMAT: 'Format of the Mat objects returned by retrieve()',
    cv2.CAP_PROP_MODE: 'Backend-specific value indicating the current capture mode',
    cv2.CAP_PROP_BRIGHTNESS: 'Brightness of the image (only for cameras)',
    cv2.CAP_PROP_CONTRAST: 'Contrast of the image (only for cameras)',
    cv2.CAP_PROP_SATURATION: 'Saturation of the image (only for cameras)',
    cv2.CAP_PROP_HUE: 'Hue of the image (only for cameras)',
    cv2.CAP_PROP_GAIN: 'Gain of the image (only for cameras)',
    cv2.CAP_PROP_EXPOSURE: 'Exposure (only for cameras)',
    cv2.CAP_PROP_AUTO_EXPOSURE: 'Boolean flag indicating whether automatic exposure is enabled',
    cv2.CAP_PROP_SHARPNESS: 'Sharpness of the image (only for cameras)',
    cv2.CAP_PROP_BACKLIGHT: 'Backlight compensation (only for cameras)',
    cv2.CAP_PROP_AUTOFOCUS: 'Boolean flag indicating whether automatic focus is enabled'
}

# Open the first camera
cap = cv2.VideoCapture(0)

# Print all the available properties
print("Properties for Camera 1:")
for prop_id in prop_id_to_name.keys():
    prop_val = cap.get(prop_id)
    print(prop_id_to_name[prop_id], prop_val)

# Open the second camera
cap = cv2.VideoCapture(2)

# Print all the available properties
print("Properties for Camera 2:")
for prop_id in prop_id_to_name.keys():
    prop_val = cap.get(prop_id)
    print(prop_id_to_name[prop_id], prop_val)
```

```python
import turtle
import numpy as np
import cv2
import random
import pygame

def init_cameras(id1=0, id2=1):
    cap1 = cv2.VideoCapture(id1)
    cap2 = cv2.VideoCapture(id2)
    cv2.namedWindow('Webcam 1')
    cv2.namedWindow('Webcam 2')
    return cap1, cap2
    
def get_frame():
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    cv2.imshow('Webcam 1', frame1)
    cv2.imshow('Webcam 2', frame2)
    return (frame1, frame2)

def setup_turtle(width, heigth, speed=1):
    turtle.setup(width, heigth)
    turtle.clearscreen()
    turtle.setworldcoordinates(0, 0, turtle.window_width(), turtle.window_height())
    turtle.speed(speed)
    turtle.goto(0, 0)    

def setup_turtle(width, heigth, speed=1):  
    screen = turtle.Screen()     # Create a window object
    screen.setup(width, heigth)    # Set the width of the window
    screen.setworldcoordinates(0, 0, turtle.window_width(), turtle.window_height())
    screen.delay(speed)    # Setup free-delay painting
    return screen

def create_points(n_points, window_width, window_heigth):
    points = [] 
    x1 = np.linspace(0, window_width, n_points, dtype = 'int')
    y1 = np.array([0, window_heigth]*(n_points//2))
    x2 = np.array([0, window_width]*(n_points//2))
    y2 = np.linspace(window_heigth, 0, n_points, dtype = 'int')
    points = list(zip(np.append(x1,x2), np.append(y1,y2)))

    return points

def move_to_points(list_of_points, n, capture = False):
    
    data = []
    for ((sx, sy),(ex,ey)) in zip(list_of_points[:-1], list_of_points[1:]):
        for x, y in zip(np.linspace(sx, ex, n, dtype='int'), np.linspace(sy, ey, n, dtype='int')):
            x, y = int(x), int(y)
            turtle.goto(x,y)
            if capture:
                frame1, frame2 = get_frame()
            else:
                frame1, frame2 = [],[]
            data.append({'x':x, 'y':y, 'camera1':frame1, 'camera2':frame2})       
    return data
```

```python
import pygame

# initialize Pygame
pygame.init()

# set up the screen
screen = pygame.display.set_mode((0, 0), pygame.FULLSCREEN)
width, height = screen.get_size()

# set up the clock
clock = pygame.time.Clock()

# set up the circle
circle_radius = 5
circle_color = (0, 0, 0)  # black
circle_pos = [0, 0]
circle_speed = 5

# main loop
while True:
    # handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()
    
    # move the circle
    circle_pos[0] += circle_speed
    circle_pos[1] += circle_speed
    
    # bounce off the right edge
    if circle_pos[0] + circle_radius > width:
        circle_pos[0] = width - circle_radius
        circle_speed = -circle_speed
    
    # bounce off the bottom edge
    if circle_pos[1] + circle_radius > height:
        circle_pos[1] = height - circle_radius
        circle_speed = -circle_speed
    
    # clear the screen
    screen.fill((255, 255, 255))  # white
    
    # draw the circle
    pygame.draw.circle(screen, circle_color, circle_pos, circle_radius)
    
    # update the screen
    pygame.display.update()
    
    # wait for a fixed amount of time
    clock.tick(60)
```
